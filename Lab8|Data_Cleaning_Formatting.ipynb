# Exercise 1: Cleaning Column Names
To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

Hint:

Column names should be in lower case
White spaces in column names should be replaced by _
st could be replaced for state

```python
#First we need to define a function to standarize the column names

def standardize_column_name(column_name):
    return column_name.lower().replace(' ', '_')

# Then we can applying the standardization function to all column names of dataframe

df.columns = [standardize_column_name(column) for column in df.columns]
df.head(10)
```

#  In the above example, we're converting names to lowercase and replacing spaces with underscores.
 

```python
# Replacing the 'st' column as 'state'

# For this we are defining a dictionary to specify the replacements
column_replacements = {
    'st': 'state'
}

# Using lambda function we can replace the column names: 
df.rename(columns=lambda x: x.replace('_st', '_state'), inplace=True)
df.rename(columns=column_replacements, inplace=True)

df.head(10)

```

# Step 3: Implement the Strategy
For example filling null values in numerical columns with the mean and in categorical columns with the mode

```python
# Handling null values
#Fill numeric columns with mean
numeric_columns = ['customer_lifetime_value', 'number_of_open_complaints']
df_copy[numeric_columns] = df_copy[numeric_columns].fillna(df[numeric_columns].mean())


# Verify that null values have been handled
null_counts_after_handling = df_copy.isnull().sum()
print("Null Values After Handling:")
print(null_counts_after_handling)
```

```python
# Fill categorical columns with mode
categorical_columns = ['state', 'policy_type']
df_copy[categorical_columns] = df_copy[categorical_columns].fillna(df_copy[categorical_columns].mode().iloc[0])

# Verify that null values have been handled
null_counts_after_handling = df_copy.isnull().sum()
print("Null Values After Handling:")
print(null_counts_after_handling)
```
