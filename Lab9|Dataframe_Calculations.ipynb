# 8. Create a new column called Combo Type whose value combines Type 1 and Type 2.
Rules:

If both Type 1 and Type 2 have valid values, the Combo Type value should contain both values in the form of <Type 1> <Type 2>. For example, if Type 1 value is Grass and Type 2 value is Poison, Combo Type will be Grass-Poison.

If Type 1 has valid value but Type 2 is not, Combo Type will be the same as Type 1. For example, if Type 1 is Fire whereas Type 2 is NaN, Combo Type will be Fire.

```python
# First we need to define function to create the "Combo Type" based on the rules
def create_combo_type(row):
    # Condition1: Check if both "Type 1" and "Type 2" have valid values
    if pd.notna(row['Type 1']) and pd.notna(row['Type 2']):
        return f"{row['Type 1']}-{row['Type 2']}"
    # Condition2: If only "Type 1" has a valid value
    elif pd.notna(row['Type 1']):
        return row['Type 1']
    # If both "Type 1" and "Type 2" are NaN, return NaN
    else:
        return np.nan

# Apply the defined function to create the "Combo Type" column
pokemon_df['Combo Type'] = pokemon_df.apply(create_combo_type, axis=1)

# Display the DataFrame with the new "Combo Type" column
pokemon_df.sample(10)
```

```python
# using numpy.nan to represent NaN values in the "Type 2" column.
# Defining custom function create_combo_type that takes a row as input and applies the conditions/rules we have given to get 
#the 'Combo type'
# Using the apply() method along with the custom function to create the "Combo Type" column based on the
#rules for each row.
```

Now our expectation is:

The information we need to collect is:

How to convert two string variables to numerical?
Let's address the first question first. You can use a method called One Hot Encoding which is frequently used in machine learning to encode categorical string variables to numerical. The idea is to gather all the possible string values in a categorical field and create a numerical field for each unique string value. Each of those numerical fields uses 1 and 0 to indicate whether the data record has the corresponding categorical value. A detailed explanation of One Hot Encoding can be found in this article. You will formally learn it in Module 3.

For instance, if a pokemon has Type 1 as Poison and Type 2 as Fire, then its Poison and Fire fields are 1 whereas all other fields are 0. If a pokemon has Type 1 as Water and Type 2 as NaN, then its Water field is 1 whereas all other fields are 0.

In the next cell, use One Hot Encoding to encode Type 1 and Type 2. Use the pokemon type values as the names of the numerical fields you create.

```python

# Use One Hot Encoding to encode Type 1 and Type 2
type1_encoded = pd.get_dummies(pokemon_df['Type 1'], prefix='Type')
type2_encoded = pd.get_dummies(pokemon_df['Type 2'], prefix='Type')
```

```python
# Concatenate the encoded columns with the original dataset
pokemon_df_encoded = pd.concat([pokemon_df, type1_encoded, type2_encoded], axis=1)

# Display the first few rows of the encoded dataset
display(pokemon_df_encoded.head())
```

```python
# Using pd.get_dummies function to perform One Hot Encoding on both the "Type 1" and "Type 2" columns.

# Prefix parameter is used to add a prefix to the newly created columns, making it clear which type each column represents.

# Then concatenate the encoded columns with the original dataset to create the pokemon_data_encoded DataFrame.
```

# "Identify VIP and Preferred Customers" is the non-technical goal of your boss. You need to translate that goal into technical languages that data analysts use:

# How to label customers whose aggregated amount_spent is in a given quantile range?
We break down the main problem into several sub problems:

Sub Problem 1: How to aggregate the amount_spent for unique customers?

Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?

Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.

```python
# Calculate the aggregated expenses for each customer
customer_expenses = orders_df.groupby('customerid')['amount_spent'].sum().reset_index()
print(customer_expenses)
```

```python
# Calculate the 75th and 95th percentiles
percentiles = customer_expenses['amount_spent'].quantile([0.75, 0.95])
display(percentiles)
```

```python
# Define a function to categorize customers
def categorize_customer(expense):
    if expense > percentiles[0.95]:
        return 'VIP'
    elif expense >= percentiles[0.75]:
        return 'Preferred'
    else:
        return 'Regular'

# Apply the categorize_customer function to create a new 'CustomerCategory' column
customer_expenses['CustomerCategory'] = customer_expenses['amount_spent'].apply(categorize_customer)

# Display the first few rows of the customer_expenses DataFrame
display(customer_expenses.head())
```

```python
# Calculate the aggregated expenses for each customer.
# Determine the 75th and 95th percentiles for aggregated expenses.
# Assign customer categories based on their expenses.
```

# Q2: How to identify which country has the most VIP Customers?

```python
# Merge the customer_expenses DataFrame with the main dataset to get customer countries
merged_df = orders_df.merge(customer_expenses, on='customerid')
```

```python
# Count the VIP Customers by country
vip_customers_by_country = merged_df[merged_df['CustomerCategory'] == 'VIP'].groupby('country')['customerid'].nunique()
print(vip_customers_by_country)
```

```python
# Find the country with the most VIP Customers
most_vip_country = vip_customers_by_country.idxmax()
vip_customer_count = vip_customers_by_country.max()

print(f"The country with the most VIP Customers is {most_vip_country} with {vip_customer_count} VIP Customers.")
```

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# Count VIP+Preferred Customers by country
vip_preferred_customers_by_country = merged_df[(merged_df['CustomerCategory'] == 'VIP') | (merged_df['CustomerCategory'] == 'Preferred')].groupby('country')['customerid'].nunique()
print(vip_preferred_customers_by_country)
```

```python
# Find the country with the most VIP+Preferred Customers combined
most_vip_preferred_country = vip_preferred_customers_by_country.idxmax()
vip_preferred_customer_count = vip_preferred_customers_by_country.max()

print(f"The country with the most VIP+Preferred Customers combined is {most_vip_preferred_country} with {vip_preferred_customer_count} customers.")
```
