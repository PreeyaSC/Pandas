# For my own purpose and to remember writing some terms used for above plot and for the future plots

#1. fig size : "figsize" stands for "figure size." It is a parameter that you can specify when creating a figure using the plt.subplots() function or when saving a plot to a file. The figsize parameter allows us to control the dimensions (width and height) of the figure in inches.figsize is a tuple of two values: (width, height).
width specifies the width of the figure in inches.
height specifies the height of the figure in inches.

#2. plt.xticks: In the code provided, plt.xticks is used to customize the x-axis tick labels in the Matplotlib plots. plt.xticks is a function that sets the properties of the x-axis ticks. For example for above code:

range(12): provides the positions (0 to 11) where the ticks will be placed on the x-axis. This corresponds to the 12 months.

calendar.month_abbr[1:]:  provides the labels for the ticks. calendar.month_abbr is a list of month abbreviations provided by the calendar module, and [1:] slices the list to exclude the empty string at the beginning (index 0), so it starts from January (index 1).

#4.plt.xticks(rotation=0): The line plt.xticks(rotation=0) is used to set the rotation angle of the x-axis tick labels to 0 degrees in the Matplotlib plot.In some cases, when the x-axis tick labels (such as text labels for months or long category names) are too long or overlap with each other, it can make the plot less readable. Rotating the tick labels can be a solution to make them fit better within the available space and prevent overlap.However, in the specific context of the code provided, the x-axis tick labels are already quite short (e.g., "Jan" for January), and there's enough space for them to be displayed horizontally without overlapping.

#3.plt.sca(axes[0]):The line plt.sca(axes[0]) is used to set the current active axis (subplot) to the first subplot in the Matplotlib figure. This allows you to specify which subplot you want to work on when you have multiple subplots within the same figure.

#4.plt.sca(axes[1]):The line plt.sca(axes[1]) is similar to plt.sca(axes[0]), but it sets the current active axis (subplot) to the second subplot in the Matplotlib figure (axes[1]).

. Minutes Lightly Active vs Steps

```python

# List of feature pairs (feature_name, x_label)
feature_pairs = [
    ('minutes lightly active', 'minutes lightly active vs steps'),
    ('minutes very active', 'minutes very active vs steps'),
    ('minutes sedentary', 'minutes sedentary vs steps')
]

# Create a loop to plot the scatter plots
for feature_name, x_label in feature_pairs:
    plt.scatter(fitbit_df[feature_name], fitbit_df['steps'], alpha=0.5)
    plt.xlabel(feature_name)
    plt.ylabel('steps')
    plt.title(x_label)
    plt.grid(True)
    plt.show()


```

# 
>In the above code,alpha=0.5 is used to control the transparency of the points in the scatter plot

>plt.grid(True) adds a grid to the plot for better visualization.

>feature_pairs is a list of tuples where each tuple contains the feature name and the desired x-axis label for the scatter plot.

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Define the number of bins for each subplot
bins_values = [10, 20, 50]

# Create subplots for each bin value
for i, bins in enumerate(bins_values):
    axes[i].hist(titanic_df['Age'], bins=bins, edgecolor='k', alpha=0.7)
    axes[i].set_xlabel('Age')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Distribution of Age (Bins={bins})')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

# How does the bin size affect your plot?

# 
The bin size in a histogram affects the level of detail and granularity in visualizing the distribution of data.

In the case of the histogram plots of age distribution with different bin sizes:
1. A larger bin size (e.g., 50) provides a broad overview of age distribution but may miss finer details.
2. A moderate bin size (e.g., 20) strikes a balance between granularity and simplicity.
3. A smaller bin size (e.g., 10) offers a detailed view of age distribution, revealing more specific patterns and variations.
The choice of bin size depends on the specific analysis objectives and the characteristics of the dataset you're working with.
